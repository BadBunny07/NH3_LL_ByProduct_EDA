## Analysing Missing values

```python
##the list of features which has missing values
features_with_nan=[features for features in dataset.columns if dataset[features].isnull().sum()>1]

for feature in features_with_nan:
    print(feature, np.round(dataset[feature].isnull().mean(),2), '% missing values')
```

## Finding relation between missing values and Availability

```python
for feature in features_with_nan:
    data=dataset.copy()
    
    data[feature]=np.where(data[feature].isnull(),1,0)
    
    data.groupby(feature)['SI_HM'].median().plot.bar()
    plt.title(feature)
    plt.show()
      
    
```

# Continuous Variables

```python
cont_features=[features for features in num_features if features not in discrete_feature]
print("Continuous features count {}".format(len(cont_features)))
```

```python
# analysing continuous features by creating histograms to understand the distribution

for feature in cont_features:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

```python
## using logarithmic transformation 

for feature in cont_features:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        #data[feature]=np.log(data[feature])    
        #data['SI_HM']=np.log(data['SI_HM'])
        plt.scatter(data[feature],data['SI_HM'])
        plt.xlabel(feature)
        plt.ylabel('SI_HM')
        plt.title(feature)
        plt.show()    
        
plt.scatter(data['CALCULATED_AVAILABILITY'],data['SI_HM'])
plt.xlabel('CALCULATED_AVAILABILITY')
plt.ylabel('SI_HM')
plt.title('CALCULATED_AVAILABILITY')
```

### Outliers

```python
for feature in cont_features:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        #data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()

data.boxplot(column='CALCULATED_AVAILABILITY')
plt.ylabel('CALCULATED_AVAILABILITY')
plt.title('CALCULATED_AVAILABILITY')
plt.show()
```
