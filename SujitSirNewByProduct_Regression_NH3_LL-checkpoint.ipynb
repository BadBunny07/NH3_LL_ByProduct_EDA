```python
#function for linear regression-#
def lin_reg(df_train,df_val):
    #fix seed
    random.seed(42)
    #provide data
    x=df_train.drop('FREE_AMMN_PPM_LL',1)
    y=df_train['FREE_AMMN_PPM_LL']
    #Create a model and fit it
    model = LinearRegression().fit(x, y)
    #Get results
    r_sq = model.score(x, y)
    #print('coefficient of determination:', r_sq)
    #adjusted r squared
    adj_r_sq=1 - (1-model.score(x, y))*(len(y)-1)/(len(y)-x.shape[1]-1)
    #print('adjusted r squared:', adj_r_sq)
    #intercept
    intercept=model.intercept_
    #print('intercept:', model.intercept_)
    #coefficient
    coefficient=model.coef_
    #print('coefficient:', model.coef_)
    #Predict response
    y_pred = model.predict(df_val.drop(['FREE_AMMN_PPM_LL'],1))    
    #print('predicted response:', y_pred, sep='\n')
    #residual
    residual=df_val['FREE_AMMN_PPM_LL']-y_pred
    #print('residual:', residual, sep='\n')
    #mae
    mae = mean_absolute_error(df_val['FREE_AMMN_PPM_LL'], y_pred)
    #print('mae:', mae, sep='\n')
    #mse
    mse = mean_squared_error(df_val['FREE_AMMN_PPM_LL'], y_pred)
    #print('mse:', mse, sep='\n')
    #rmse
    rmse = math.sqrt(mse)
    #print('rmse:', rmse, sep='\n')
    #mape
    mape = np.mean(np.abs((df_val['FREE_AMMN_PPM_LL'] - y_pred)/df_val['FREE_AMMN_PPM_LL']))*100
    #print('rmse:', rmse, sep='\n')
    #Test R squre
    test_r2 = r2_score(df_val['FREE_AMMN_PPM_LL'], y_pred) 
    #print('r2 score for test data is', test_r2) 
    #Strike Rate Calculation#
    sr = len(residual[abs(residual)<=1000])/len(residual)
#     print('strike rate for test data is', sr)
    return y_pred,residual,r_sq,intercept,coefficient,x,sr,mae,mse,rmse,test_r2,adj_r_sq,mape,model
```

```python
#function call--store actual vs predicted result--#
#fix seed#
for i in df_val.columns:    
    random.seed(3)
    y_pred,residual,r_sq,intercept,coefficient,x,sr,mae,mse,rmse,test_r2,adj_r_sq,mape,model=lin_reg(df_train,df_val) #insert validation data
    df1=pd.DataFrame(columns=['Time_Stamp','Actual','Predicted','Residual'])
    df1['Time_Stamp']=df_val[i].index
    df1['Actual']=df_val['FREE_AMMN_PPM_LL'].to_list()
    df1['Predicted']=y_pred
    df1['Residual']=residual.to_list()
#write the output at desired location#
df1.to_excel(r'nh3_ll_actual_vs_predicted_v5_unseen_4_raw_Try1.xlsx',index=False)
```

```python
#coefficient relation with features#
for i in df_train.columns:
    #fix seed#
    random.seed(3)
    y_pred,residual,r_sq,intercept,coefficient,x,sr,mae,mse,rmse,test_r2,adj_r_sq,mape,model=lin_reg(df_train,df_val) #insert validation data
    coeff=pd.DataFrame(coefficient,columns=['coefficient'],index=x.columns)
    #reset index#
    coeff=coeff.reset_index()
    #rename column#
    coeff.rename(columns={'index':'Feature'},inplace=True)
coeff
```

```python
#function call--store trained result--#
mae_=[]
mse_=[]
rmse_=[]
mape_=[]
rsq=[]
tst_rsq=[]
inter=[]
feat=[]
sr_=[]
for i in x.columns:
    #fix seed
    random.seed(3)
    y_pred,residual,r_sq,intercept,coefficient,x,sr,mae,mse,rmse,test_r2,adj_r_sq,mape,model=lin_reg(df_train,df_val) #insert validation data
    feat.append(i)
    rsq.append(r_sq)
    sr_.append(sr)
    tst_rsq.append(test_r2)
    inter.append(intercept)
    mae_.append(mae)
    mse_.append(mse)
    rmse_.append(rmse)
    mape_.append(mape)
#Store unique result for each feator    
output=pd.DataFrame(columns=['Feature','Intercept','R_Sq','Strike_Rate','Test_R_Sq','MAE','MSE','RMSE','Mape'])
output['R_Sq']=rsq
output['Strike_Rate']=sr_
output['Test_R_Sq']=tst_rsq
output['Intercept']=inter
output['MAE']=mae_
output['MSE']=mse_
output['RMSE']=rmse_
output['Mape']=mape_
output['Feature']=feat
```

```python
#merge these 2 dataframe
result_trained=coeff.merge(output, how='left', on=['Feature']).reset_index(drop=True)
result_trained.to_excel(r'nh3_ll_output_v5_unseen_4_raw_Try1.xlsx',index=False)
result_trained
```
